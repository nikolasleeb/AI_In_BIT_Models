```python
import re # imports the pythong Regular Expression library useful for text cleaning and preprocessing
text = re.sub(r'"', '', text) # removes all double quotes from text string
text = re.sub(r'3', '', text) # removes all occurrences of number 3 from text string # not used in my example
text = re.sub(r'4', '', text) # removes all occurrences of number 4 from text string # not used in my example
text = re.sub(r'\[\]', '', text) # removes all empty square brackets from text string
text = ''.join([char for char in text if char not in string.punctuation])  # Removes all punctuation from the text string # not part of original code
```

```python
import nltk # import the Natural Language ToolKit for working with human language data

nltk.download('punkt_tab') # downloads punkt tokenizer, a pretrained model for splitting text into sentences and words

from nltk import sent_tokenize # imports function to split text into sentences
from nltk import word_tokenize # imports function to split text into words

sentences = sent_tokenize(text) # uses sent_tokenizes to tokenize cleaned text into sentences
```

```python
print (len(sentences)) # prints the number of sentences in the senteces list
```

```python
for index, sentence in enumerate(sentences): # loops through the text seperating out each sentence and assigning it an index number
  print(f"Sentence number: {index+1}")
  print(sentence)
```

```python
words = word_tokenize(text) # tokenizes cleaned text into individual words
```

```python
print (len(words)) # prints total number of words in text
```

```python
print(words) # prints the list of every words
```

```python
from nltk.probability import FreqDist # imports the Frequency Distribution class from NLTK to calculate frequency distribution of words
```

```python
fdist = FreqDist(words) # creates an objects for tokenized word list to count occurrences of each unique word
```

```python
print(fdist.most_common(15)) # prints the 15 most common words and their counts
```

```python
from nltk.corpus import stopwords # imports the stopwords module form NLTK which provides llist of common stop words(this, is, and, or, etc)
nltk.download('stopwords') # downloads the stop words dataset

stop_words = set(stopwords.words('english')) # creates a set of stop words for english language
filtered_words = [w for w in words if not w.lower() in stop_words] # uses list comprehension to iterate through the words list and includes words not in stop words list
```

```python
print(filtered_words) # outputs the words after removing the commong stop words
```

```python
import string # imports the string module to remove punctuation

no_punct = [w for w in filtered_words if w not in string.punctuation] # removes the punctuation # i added this code above

print(no_punct) # prints the list with no punctuation
```

```python
fdist = FreqDist(no_punct) # counts the most common words after removing punctuation
```

```python
print(fdist.most_common(5)) # prints the 5 most common words and thier count
```

```python
fdist.plot(12) # plots the 12 most common words
plt.xlabel("Words") # adds word label
plt.ylabel("Word Count") # adds word count label
plt.show() # shows the plot
```

```python
from nltk import Text # imports text class from NLTK to work with tokenized text as an object

cleanText = Text(no_punct) # init a text object using the cleaned list of words

word_freq = dict([(word, cleanText.count(word)) for word in set(cleanText)]) # creates a dictionary where unique words in cleantext object are shows as frequenices
for word, count in word_freq.items(): print(f"{word:<15} {count:>10}") # prints each word on the left and its count on the right
```
